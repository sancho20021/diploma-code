```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_seq = pd.read_csv('parallel_graphs/sequential.csv')
df_par = pd.read_csv('parallel_graphs/parallel.csv')

graph_width = 5
graph_depth = 5
nirvana_bandwidth = 5 * 200

fig, axs = plt.subplots(nrows=3, figsize=(17, 14))
axs = {
    'min_iters': axs[0],
    'max_iters': axs[1],
    'sum_iters': axs[2],
}
for ax in axs.values():
    ax.set_xlabel('graphs batch number')
    ax.set_title(f'Differences in sequential / parallel execution of nirvana graph batch. {graph_width=}, {graph_depth=}, {nirvana_bandwidth=}')
    ax.grid()

axs['min_iters'].set_ylabel('min iters')
axs['min_iters'].plot(df_seq['graphs_number'], df_seq['min_iters'], color='red', label='sequential execution')
axs['min_iters'].plot(df_par['graphs_number'], df_par['min_iters'], color='green', label='parallel execution')

axs['max_iters'].set_ylabel('max iters')
axs['max_iters'].plot(df_seq['graphs_number'], df_seq['max_iters'], color='red', label='sequential execution')
axs['max_iters'].plot(df_par['graphs_number'], df_par['max_iters'], color='green', label='parallel execution')

axs['sum_iters'].set_ylabel('sum iters')
axs['sum_iters'].plot(df_seq['graphs_number'], df_seq['sum_iters'], color='red', label='sequential execution')
axs['sum_iters'].plot(df_par['graphs_number'], df_par['sum_iters'], color='green', label='parallel execution')

for ax in axs.values():
    ax.legend()
```
